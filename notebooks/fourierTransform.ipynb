```python
import librosa
import librosa.display

import IPython.display as ipd
import matplotlib.pyplot as plt

import numpy as np
import scipy as sp
```

```python
piano_c = "../sampleAudio/piano_c.wav"
```

```python
ipd.Audio(piano_c)
```

```python
c_sig, c_sr = librosa.load(piano_c)
```

```python
librosa.display.waveshow(c_sig,sr=c_sr)
plt.show()
```

$$
A\sin(kx - \omega t + \phi) \\
\text{or} \\
A\sin(2\pi ft + \phi)
$$

```python
samples = range(len(c_sig))
t = librosa.samples_to_time(samples, sr=c_sr)

f = 523
phase = 0.55
sin = 0.1*np.sin(2*np.pi*(f*t - phase))
plt.figure(figsize=(18, 8))

plt.plot(t[10000:10400], c_sig[10000:10400]) 
plt.plot(t[10000:10400], sin[10000:10400], color="r")

plt.fill_between(t[10000:10400], sin[10000:10400]*c_sig[10000:10400], color="y")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# plot spectrum
ft = sp.fft.fft(c_sig)
magnitude = np.absolute(ft)
frequency = np.linspace(0, c_sr, len(magnitude)) 
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], magnitude[:5000]) # magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# Calculate phase spectrum
phase_spectrum = np.angle(ft)
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], phase_spectrum[:5000])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
plt.show()
```
